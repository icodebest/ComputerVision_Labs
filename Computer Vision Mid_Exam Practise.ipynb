```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
#load the image
image1=cv2.imread("/content/Screenshot 2025-04-16 232431.png")
```

```python
image_rgb=cv2.cvtColor(image1, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(image_rgb)
plt.axis("off")
plt.show()
```

```python
#convert image to grayscale
gray_image=cv2.cvtColor(image1, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(gray_image, cmap="gray")
plt.axis("off")
plt.show()
```

```python
#resize the image
resized_image=cv2.resize(image1, (300,300))
resized_rgb=cv2.cvtColor(resized_image, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(resized_rgb)
plt.axis("off")
plt.show()
```

```python
#flip an image
flipped_hori=cv2.flip(image1, 1)
flip_vert=cv2.flip(image1, 0)

flip_hori_rgb=cv2.cvtColor(flipped_hori, cv2.COLOR_BGR2RGB)
flip_vert_rgb=cv2.cvtColor(flip_vert, cv2.COLOR_BGR2RGB)
```

```python
#subplot code
fig, ax=plt.subplots(1,3, figsize=(20, 10))

ax[0].imshow(image_rgb)
ax[0].set_title("Original Image")
ax[0].axis("off")

ax[1].imshow(flip_hori_rgb)
ax[1].set_title("Horizental flipped")
ax[1].axis("off")

ax[2].imshow(flip_vert_rgb)
ax[2].set_title("Vertically flipped")
ax[2].axis("off")

plt.show()
```

```python
#contast stretching
image_gray=cv2.imread("/content/Screenshot 2025-04-16 232431.png", cv2.IMREAD_GRAYSCALE)
```

```python
#apply contrast streteching
min_pixel=np.min(image_gray)
max_pixel=np.max(image_gray)
```

```python
stretched_image=(image_gray-min_pixel)/(max_pixel-min_pixel)*255
```

```python
# Display images
fig, ax = plt.subplots(1, 2, figsize=(10, 5)) # make sublot for original image and the streteched image
ax[0].imshow(image_gray, cmap='gray') # show the original image
ax[0].set_title("Original Image") # show title
ax[0].axis("off")

ax[1].imshow(stretched_image, cmap='gray') # for the streteched image
ax[1].set_title("Contrast Stretched Image") # Give title to athat one
ax[1].axis("off") # off the axis of the image

plt.show() # plot the image
```

### Image_smoothing

```python
#Gaussian Blur
blurred=cv2.GaussianBlur(image1, (5,5), 0)
```

```python
fig, axis=plt.subplots(1,2, figsize=(10,5))
axis[0].imshow(image1)
axis[0].set_title("Original image")
axis[0].axis("off")

axis[1].imshow(blurred)
axis[1].set_title("Gaussian blured image")
axis[1].axis("off")

plt.show()
```

```python
#image sharpening
sharpen_kernal=np.array([[0,-1,0],
                         [-1,5,-1],
                         [0,-1,0]])
```

```python
sharpened=cv2.filter2D(image1,-1, sharpen_kernal)

# Display images
fig, axis= plt.subplots(1,2, figsize=(10,5))
axis[0].imshow(image1)
axis[0].set_title("Original Image")
axis[0].axis("off")

axis[1].imshow(sharpened)
axis[1].set_title("Sharpened Image")
axis[1].axis("off")

plt.show()
```

```python
gray=cv2.cvtColor(image1,cv2.COLOR_BGR2GRAY)
#canny edge dtection
edges=cv2.Canny(gray, 50,150)
```

```python
# Display images
fig, axis= plt.subplots(1,2, figsize=(10,5))
axis[0].imshow(gray, cmap='gray')
axis[0].set_title("Grayscale Image")
axis[0].axis("off")

axis[1].imshow(edges, cmap='gray')
axis[1].set_title("Edge Detection (Canny)")
axis[1].axis("off")

plt.show()
```

```python
#median blur
median_blurred=cv2.medianBlur(image1, 5)
#Display images
fig, axis = plt.subplots(1, 2, figsize=(10, 5))
axis[0].imshow(image1)
axis[0].set_title("Original Image")
axis[0].axis("off")

axis[1].imshow(median_blurred)
axis[1].set_title("Median Blurred Image")
axis[1].axis("off")

plt.show()
```

```python
sharpen_kernal=np.array([[0,-1,0],
                         [-1,6,-1],
                         [0,-1,0]])
sharpened=cv2.filter2D(image1,-1, sharpen_kernal)

# Display images
fig, axis= plt.subplots(1,2, figsize=(10,5))
axis[0].imshow(image1)
axis[0].set_title("Original Image")
axis[0].axis("off")

axis[1].imshow(sharpened)
axis[1].set_title("Sharpened Image")
axis[1].axis("off")

plt.show()
```

```python
# Function to display an image using Matplotlib
def show_image(title, image):
    plt.figure(figsize=(5, 5))
    plt.imshow(image, cmap='gray')
    plt.title(title)
    plt.axis('off')
    plt.show()
```

```python
_,binary_image=cv2.threshold(image1, 127,255,cv2.THRESH_BINARY )

#define a kernal
kernal=np.ones((5,5), np.uint8)

#apply dilation
dilation=cv2.dilate(binary_image, kernal, iterations=1)

#apply erosion
erosion=cv2.erode(binary_image, kernal, iterations=1)

#display results
show_image("Binary Image", binary_image)
show_image("Dilation", dilation)
show_image("Erosion", erosion)
```
