## Question No :01
Implement different image filtering techniques (Gaussian, median, and bilateral
filtering) using OpenCV. Compare their effects on noisy images and explain which filter is most effective for different types of noise.

```python
#import libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('/content/image1.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
# Apply Gaussian blur
gaussian_blur = cv2.GaussianBlur(img, (5, 5), 0)
```

```python
# Apply Median blur
median_blur = cv2.medianBlur(img, 5)
```

```python
# Apply Bilateral filter
bilateral_blur = cv2.bilateralFilter(img, 9, 75, 75)
```

```python
# Display the original and filtered images
plt.figure(figsize=(15, 10))

plt.subplot(2, 2, 1)
plt.imshow(img)
plt.title('Original Image')

plt.subplot(2, 2, 2)
plt.imshow(gaussian_blur)
plt.title('Gaussian Blur')

plt.subplot(2, 2, 3)
plt.imshow(median_blur)
plt.title('Median Blur')

plt.subplot(2, 2, 4)
plt.imshow(bilateral_blur)
plt.title('Bilateral Blur')

plt.show()
```

Gaussian Blur: Smooths everything, reduces overall noise.

Median Blur: Smooths while protecting edges better than Gaussian.
Bilateral Blur: Smooths while preserving edges best, but might be computationally more intensive.

## Question No: 02
Write a Python program on an image to perform morphological operations
(Opening, Closing, Morphological Gradient, Top-Hat, and Black-Hat). How do these operations help in enhancing different image features?

```python
img = cv2.imread('/content/image1.jpg', cv2.IMREAD_GRAYSCALE)  # Load as grayscale
```

```python
 # Define a kernel
kernel = np.ones((5, 5), np.uint8)
```

```python
# Opening: Erosion followed by dilation
opening = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel)
```

```python
# Closing: Dilation followed by erosion
closing = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel)
```

```python
# Morphological Gradient: Difference between dilation and erosion
gradient = cv2.morphologyEx(img, cv2.MORPH_GRADIENT, kernel)
```

```python
# Top-Hat: it is Original image minus opening (Highlights bright regions on a dark background)
tophat = cv2.morphologyEx(img, cv2.MORPH_TOPHAT, kernel)
```

```python
 # Black-Hat: Closing minus original image (Highlights dark regions on a bright background)
blackhat = cv2.morphologyEx(img, cv2.MORPH_BLACKHAT, kernel)
```

```python
# Display the results
plt.figure(figsize=(15, 10))

plt.subplot(2, 3, 1)
plt.imshow(img, cmap='gray')
plt.title('Original Image')

plt.subplot(2, 3, 2)
plt.imshow(opening, cmap='gray')
plt.title('Opening')

plt.subplot(2, 3, 3)
plt.imshow(closing, cmap='gray')
plt.title('Closing')

plt.subplot(2, 3, 4)
plt.imshow(gradient, cmap='gray')
plt.title('Morphological Gradient')

plt.subplot(2, 3, 5)
plt.imshow(tophat, cmap='gray')
plt.title('Top-Hat')

plt.subplot(2, 3, 6)
plt.imshow(blackhat, cmap='gray')
plt.title('Black-Hat')

plt.show()
```

- Opening: Cleans up noise and isolates objects.
- Closing: Fills gaps and connects objects.
- Morphological Gradient: Highlights object edges.
- Top-Hat: Enhances bright features.
- Black-Hat: Enhances dark features.

## Question No: 03
 Implement Sobel, Laplacian, and Canny edge detection using OpenCV. Visualize
and compare the results. Which method performs best on images with fine details, and
why?

```python
# Apply Gaussian blur to reduce noise
blurred_img = cv2.GaussianBlur(img, (3, 3), 0)
```

```python
# Sobel Edge Detection
sobelx = cv2.Sobel(blurred_img, cv2.CV_64F, 1, 0, ksize=5)
sobely = cv2.Sobel(blurred_img, cv2.CV_64F, 0, 1, ksize=5)
sobel_edges = cv2.magnitude(sobelx, sobely)
sobel_edges = cv2.convertScaleAbs(sobel_edges)
```

```python
# Laplacian Edge Detection
laplacian_edges = cv2.Laplacian(blurred_img, cv2.CV_64F)
laplacian_edges = cv2.convertScaleAbs(laplacian_edges)
```

```python
# Canny Edge Detection
canny_edges = cv2.Canny(blurred_img, 50, 150)  # Adjust threshold values as needed
```

```python
# Display the results
plt.figure(figsize=(15, 10))

plt.subplot(2, 2, 1)
plt.imshow(img, cmap='gray')
plt.title('Original Image')

plt.subplot(2, 2, 2)
plt.imshow(sobel_edges, cmap='gray')
plt.title('Sobel Edge Detection')

plt.subplot(2, 2, 3)
plt.imshow(laplacian_edges, cmap='gray')
plt.title('Laplacian Edge Detection')

plt.subplot(2, 2, 4)
plt.imshow(canny_edges, cmap='gray')
plt.title('Canny Edge Detection')

plt.show()
```

- Sobel: Detects basic vertical and horizontal edges.
- Laplacian: Detects edges in all directions but can be sensitive to noise.
- Canny: A robust and widely used method that produces clean and accurate edges.

## Question No: 05
Implement the Watershed algorithm for image segmentation in Python.
Experiment with marker-based segmentation and evaluate how well it separates
overlapping objects in an image.

```python
# Load the image (replace with your image path)
img = cv2.imread('/content/image1.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
# Apply thresholding to obtain a binary image
ret, thresh = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)
```

```python
# Noise removal using morphological opening
kernel = np.ones((3, 3), np.uint8)
opening = cv2.morphologyEx(thresh, cv2.MORPH_OPEN, kernel, iterations=2)
```

```python
# Sure background area
sure_bg = cv2.dilate(opening, kernel, iterations=3)
```

```python
# Finding sure foreground area
dist_transform = cv2.distanceTransform(opening, cv2.DIST_L2, 5)
ret, sure_fg = cv2.threshold(dist_transform, 0.7 * dist_transform.max(), 255, 0)
```

```python
# Finding unknown region
sure_fg = np.uint8(sure_fg)
unknown = cv2.subtract(sure_bg, sure_fg)
```

```python
# Marker labelling
ret, markers = cv2.connectedComponents(sure_fg)
```

```python
# Add one to all labels so that sure background is not 0, but 1
markers = markers + 1
```

```python
# Now, mark the region of unknown with zero
markers[unknown == 255] = 0
```

```python
# Apply watershed algorithm
markers = cv2.watershed(img, markers)
img[markers == -1] = [255, 0, 0]  # Highlight boundaries in blue
```

```python
# Display the results
plt.figure(figsize=(15, 10))

plt.subplot(2, 2, 1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Original Image')

plt.subplot(2, 2, 2)
plt.imshow(thresh, cmap='gray')
plt.title('Threshold Image')

plt.subplot(2, 2, 3)
plt.imshow(dist_transform, cmap='gray')
plt.title('Distance Transform')

plt.subplot(2, 2, 4)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Watershed Segmentation')

plt.show()
```
