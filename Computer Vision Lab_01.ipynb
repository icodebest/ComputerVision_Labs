### Task 4: Flip an Image

```python
flipped_horiz=cv2.flip(image,1) # 1 is used for horizental flip
flipped_vert=cv2.flip(image,0) # 0 is used for vertical flip
```

```python
#convert for correct color display
flipped_horiz_rgb=cv2.cvtColor(flipped_horiz,cv2.COLOR_BGR2RGB) # Convert the horizental flipped image color to BGR to RGB
flipped_vert_rgb=cv2.cvtColor(flipped_vert,cv2.COLOR_BGR2RGB) # Convert the vertical flipped image color to BGR to RGB
```

```python
fig, ax = plt.subplots(1, 2, figsize=(10, 5)) # make a subplot for two different flipping
ax[0].imshow(flipped_horiz_rgb) # on axis 0 plot the horizental flipped image
ax[0].set_title("Horizontally Flipped") # Give the title to the image
ax[0].axis("off") # off the axis of the image
ax[1].imshow(flipped_vert_rgb) # on axis 1 plot the vertical flipped image
ax[1].set_title("Vertically Flipped") #Give the title to image
ax[1].axis("off") # off asiz
plt.show() # show the plot
```

### Task 6: Contrast Stretching

```python
#load the image
image=cv2.imread('/content/images.jpg',cv2.IMREAD_GRAYSCALE)
```

```python
# Apply contrast stretching
min_pixel=np.min(image) # apply numpy min function on the pixel of the image
max_pixel=np.max(image)# apply numpy max function on the pixel of the image
```

```python
# streteched image
stretched_image=(image-min_pixel)/(max_pixel-min_pixel)*255
```

```python
# Display images
fig, ax = plt.subplots(1, 2, figsize=(10, 5)) # make sublot for original image and the streteched image
ax[0].imshow(image, cmap='gray') # show the original image
ax[0].set_title("Original Image") # show title
ax[0].axis("off")

ax[1].imshow(stretched_image, cmap='gray') # for the streteched image
ax[1].set_title("Contrast Stretched Image") # Give title to athat one
ax[1].axis("off") # off the axis of the image

plt.show() # plot the image
```

### Task 7: Histogtram Equalization

```python
#load the image in grayscale
image=cv2.imread('/content/images.jpg')
gray_image=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY) # convert the image fro BGR to Grayscale
```

```python
# Apply histogram equalization
equalized_image=cv2.equalizeHist(gray_image)
```

```python
#Display the image
fig, ax = plt.subplots(1, 2, figsize=(10, 5))
ax[0].imshow(gray_image,cmap='gray')
ax[0].set_title("Original Grayscale image")
ax[0].axis("off")

ax[1].imshow(equalized_image, cmap='gray')
ax[1].set_title("Histogram Equalized Image")
ax[1].axis("off")

plt.show()
```
