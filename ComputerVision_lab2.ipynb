## 1. Image Smoothing (Blurring)

### 1.1 Gaussian Blur

```python
# Importing Libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the image
image=cv2.imread("/content/download.jfif")
image=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
```

```python
#Display the image
plt.imshow(image)
plt.axis("off")
plt.show()
```

```python
# Apply Gaussian Blur
blurred=cv2.GaussianBlur(image,(5,5), 0)
```

```python
#Display images
fig, axis = plt.subplots(1, 2, figsize=(10, 5))
axis[0].imshow(image)
axis[0].set_title("Original Image")
axis[0].axis("off")

axis[1].imshow(blurred)
axis[1].set_title("Gaussian Blurred Image")
axis[1].axis("off")

plt.show()
```

## 2. Image Sharpening

```python
sharpen_kernal=np.array([[0,-1,0],
                         [-1,5,-1],
                         [0,-1,0]])
```

```python
sharpened=cv2.filter2D(image,-1, sharpen_kernal)
```

```python
# Display images
fig, axis= plt.subplots(1,2, figsize=(10,5))
axis[0].imshow(image)
axis[0].set_title("Original Image")
axis[0].axis("off")

axis[1].imshow(sharpened)
axis[1].set_title("Sharpened Image")
axis[1].axis("off")

plt.show()
```

## 3. Edge Detection using Canny Filter


```python
# Load the image
image=cv2.imread("/content/download.jfif")
image=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
```

```python
# Convert to grayscale for edge detection
gray=cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
#Apply canny edge detection
edges=cv2.Canny(gray,100,200)
```

```python
# Display images
fig, axis= plt.subplots(1,2, figsize=(10,5))
axis[0].imshow(gray, cmap='gray')
axis[0].set_title("Grayscale Image")
axis[0].axis("off")

axis[1].imshow(edges, cmap='gray')
axis[1].set_title("Edge Detection (Canny)")
axis[1].axis("off")

plt.show()
```
